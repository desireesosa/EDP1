**METODO DE SEPARACION DE VARIABLES.** \\
Use el método de separación de variables para resolver: \\
$u_t(x,t)=u_{xx}(x,t)$ con $0 < x < 1$, $t>0$ \\
donde sus condiciones iniciales son:
1.   $u_x(0,t)=0$, $u_x(1,t)=0$, $t>0$
2.   $u_x(x,0)=f(x)=x$, $0 < x < 1$ \\

Para poder obtener la solucion, aplicamos el metodo de separacion de variables y despues trazamos la grafica realizando un programa en este cuaderno que ayude a visualizarla.


*Solucion.* \
Resolviendo por metodo de separacion de variables, tenemos que:\
$u_t(x,t)=u_{xx}(x,t)$ \
aplicamos entonces: \
$u(x,t)=X(x)T(t)$ \
sustimos en la ecuacion: \
$X(x)T'(t)=X''(x)T(t)$ \
obtenemos dos ecuaciones las cuales estaran asociadas a T y a X, de ahi debemos aplicar nuestras condiciones de frontera: \
$u_x(0,t)=0$, $u_x(1,t)=0$




```python
#Programa que grafique el metodo de separacion de variables de la ecuacion de calor
from pylab import *
```

Sabemos que, la solucion general de la ecuacion es: \
$u(x,t)=∑_{n=0}^{∞}C_n{cos(nπ{x})e^{-(nπ)^{2}t}}$ \
dada esta solucion, encontramos a los coeficientes

asi, obtenemos que la ***solucion final a graficar*** es: \
\
$u(x,t)=\frac{1}{2}+∑_{n=1}^{∞}\frac{2(-1)^{n}}{(nπ)^{2}}cos(nπx)e^{-(nπ)^{2}t}$

```python
#graficamos de primera mano las condiciones iniciales
def recta(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = x[i]
    return y

x = linspace(0, 1, 1000)
f = recta(x)
plot(x, f, label='$u(x,0)=x$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Condición inicial (u(x,0)=f(x))")
xlim([-0.01, 1.01])
ylim([-0.01, 1.01])
grid()
legend()
show()
```

```python
def u_0(x, N=20):  # N es el número de armónicos
    f = zeros(len(x))
    for n in range(1, N):
      f += 2*(((-1)**n-1)/(n*pi)**2)*e**(-(n*pi)**2*0)*cos(n*pi*x)
    return (1/2)+f


f_n = u_0(x)
plt.plot(x, f_n, label='$f_N(x)$ con $N=20$')
plt.plot(x, x, label='$f(x)=x$')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Aproximación de la condición inicial")
plt.xlim([-0.01, 1.1])
plt.ylim([-0.01, 1.2])
plt.grid()
plt.legend()
plt.show()
```

Aplicamos ahora para diferentes cambios armonicos:

```python
for i in range(1,20,5):
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
plt.xlim([-0.01, 1.1])
plt.ylim([-0.01, 1.2])
plt.grid()
plt.legend()
plt.show()
```

ahora, para momento inicial en t=0

```python
def u_x_t(x,t,N=50):
    f = zeros( (len(x),len(t)) )
    for i in range(len(t)):
        for n in range(1,N):
            f[:,i] += 2*(((-1)**n-1)/(n*pi)**2)*e**(-(n*pi)**2*0)*cos(n*pi*x)
    return (1/2)+f
```

```python
x = linspace(0,1,1000)
t = linspace(0,10,1000)
U = u_x_t(x,t)
plot(x,U[:,0],label='$u(x,0)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
plt.xlim([-0.01, 1.1])
plt.ylim([-0.01, 1.2])
grid()
legend()
show()
```

ahora, cuando $t\to{∞}$

```python
plot(x,U[:,0],label='$u(x,t=0)$') # Mejorar
plot(x,U[:,1],label='$u(x,t=1)$')
plot(x,U[:,2],label='$u(x,t=2)$')
plot(x,U[:,3],label='$u(x,t=3)$')
plot(x,U[:,10],label='$u(x,t=10)$')
plot(x,U[:,20],label='$u(x,t=20)$')
plt.xlabel('$x$'); plt.ylabel('$u(x,t)$');
plt.title("Evolución de la solución")
xlim([-0.01,1.01]),ylim([-0.01,1.2]),grid(),legend()
show()
```
