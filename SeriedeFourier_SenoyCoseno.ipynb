*Funcion en Coseno y Seno* \\
$f(x)=e^{2x}$ con un intervalo $0<x<1$ \\

***Funcion IMPAR***

*Realizamos el codigo para poder obtener la grafica de la serie de fourier en coseno*

```python
#Codigo que muestre la grafica de una serie de fourier en coseno y seno
from sympy import fourier_series, pi, plot, Piecewise, exp
from sympy.abc import x
import matplotlib.pyplot as plt
```

*Definimos a $f$*

```python
f = Piecewise(
    (exp(2*x), (x >= 0) & (x < 1)),
    (-exp(-2*x), (x >= -1) & (x < 0))
)

#Comenzamos a graficar
plot(f, (x,-1,1),line_color ='pink')
```

```python
s = fourier_series(f, (x, -1, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -1, 1), show=False, legend=True)
```

*Serie de Fourier:*

```python
s
```

*Truncamos a la funcion*

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
p.show()
```
