```python
#Graficar y=x**2
"Ecuacion de la Parabola"
import matplotlib.pyplot as plt #libreria parea graficar la funcion
import numpy as np #libreria para mandar llamar un numero
x=np.linspace(-10,10,100) #intervalo de la grafica (10, 100)
y=x**2 #funcion a graficar

plt.plot(x,y) #
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x^2')
plt.show()
```
