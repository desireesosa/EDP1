$Grafica$ $de$ $y=x^2$

Se importan dos librerias para poder ejecutar a la ecuacion de la parabola, en
este caso utilizamos **matplotlib.pyplot** para poder graficar y una llamada **numpy** para llamar un numero a graficar.

Se ejecuta una graficacion (impresion) y se muestra en pantalla.



```python
#Graficar y=x**2
#Ecuacion de la Parabola

import matplotlib.pyplot as plt #libreria parea graficar la funcion
import numpy as np #libreria para mandar llamar un numero
x=np.linspace(-10,10,100) #intervalo de la grafica (10, 100)
y=x**2 #funcion a graficar

#Graficacion
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafica de $y=x^2$', color='green', fontsize=15)
plt.show()

#Agregar gif
from IPython.display import Image
Image(url='https://media2.giphy.com/media/v1.Y2lkPTc5MGI3NjExYnp1NTF1dnp4dHI5NDR4ZTJ4c2JkOHJwbjRieGVzc2duc2Z2dDBvYSZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/3og0IALtj7qQJGPs8U/giphy.gif')
gif_url = 'https://media2.giphy.com/media/v1.Y2lkPTc5MGI3NjExYnp1NTF1dnp4dHI5NDR4ZTJ4c2JkOHJwbjRieGVzc2duc2Z2dDBvYSZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/3og0IALtj7qQJGPs8U/giphy.gif'
Image(url=gif_url)
```
