**Ondas Viajeras** \\
Realizar un codigo que ejecute lo siguiente: \\
Grafique la funcion $f'(x)=exp(-\frac{(x'-0.5)^{2}}{0.01})$ y suponga que esta representa una perturbacion en una cuerda.


1.   Mostrar que al hacer el cambio de variable $x'=x-ct$ la perturbacion se desplaza a la derecha.
2.   Mostrar que al hacer el cambio de variable $x'=x+ct$ la perturbacion se desplaza a la izquierda.
3.   Graficar la superposicion de las ondas viajeras de los puntos (1) y (2).




*Realizamos un codigo que grafique a la funcion dada.*

```python
#graficar funcion
import numpy as np
import matplotlib.pyplot as plt

def perturbation(x_prime):
    return np.exp(-(x_prime - 0.5)**2 / 0.01)

x_prime_values = np.linspace(0, 1, 100)
plt.plot(x_prime_values, perturbation(x_prime_values))
plt.xlabel("$x'$")
plt.ylabel("$f'(x')$")
plt.title("Perturbacion en una cuerda de f'(x)")
plt.show()
```

*Ahora, modificamos de manera que podamos cumplir los puntos (1) y (2).* \\
**Punto (1)**  Mostrar que al hacer el cambio de variable $x'=x-ct$ la perturbacion se desplaza a la derecha.


```python
#realizamos un cambio de variable
import numpy as np
import matplotlib.pyplot as plt
```

```python
#a√±adimos tiempos y velocidad de onda
c=0.1
t_values=[0,0.5,1]
```

```python
#graficamos
for t in t_values:
    x_values = np.linspace(0, 2, 200)
    x_prime_shifted_right = x_values - c * t
    plt.plot(x_values, perturbation(x_prime_shifted_right), label=f't = {t}')

plt.xlabel("$x$")
plt.ylabel("$f'(x-ct)$")
plt.title("Desplazamiento de la perturbacion a la derecha")
plt.legend()
plt.show()
```

**Punto (2)** Mostrar que al hacer el cambio de variable $x'=x+ct$ la perturbacion se desplaza a la izquierda.

```python
#realizamos un cambio de variable
import numpy as np
import matplotlib.pyplot as plt
```

```python
#agregamos tiempo y velocidad de onda
c=0.1
t_values=[0,0.5,1]
```

```python
#graficamos
for t in t_values:
    x_values = np.linspace(0, 2, 200)
    x_prime_shifted_left = x_values + c * t
    plt.plot(x_values, perturbation(x_prime_shifted_left), label=f't = {t}')

plt.xlabel("$x$")
plt.ylabel("$f'(x+ct)$")
plt.title("Desplazamiento de la perturbacion a la izquierda")
plt.legend()
plt.show()
```

*Por ultimo, realizamos el punto (3)*

```python
#Graficar la superposicion de las ondas viajeras de los puntos (1) y (2).
import numpy as np
import matplotlib.pyplot as plt
```

```python
#elegimos un tiempo y una velocidad
c=0.1
t_values=[0, 0.5, 1]
```

```python
#realizamos la superposicion
x_values = np.linspace(0, 2, 200)

def perturbation(x_prime):
    return np.exp(-(x_prime - 0.5)**2 / 0.01)

for t in t_values:
    x_prime_shifted_right = x_values - c * t
    x_prime_shifted_left = x_values + c * t
    y_superposition = 0.5 * (perturbation(x_prime_shifted_right) + perturbation(x_prime_shifted_left))
    plt.plot(x_values, y_superposition, label=f't = {t} (superposici√≥n)', linestyle=':', linewidth=2)

plt.xlabel("$x$")
plt.ylabel("Superposici√≥n")
plt.title("Superposici√≥n de ondas viajeras")
plt.legend()
plt.show()
```
